#### Penanganan 'description'

```python
movies['description']
```

```python
# menghapus baris yang memiliki null
# parameter inplace => menimpa data yang sebelumnya jika True
movies.dropna(subset=['description'], inplace=True)
movies.isna().sum()
```

#### Penanganan 'genre'

```python
# Menangani missing value di kolom genre
# Karena genre bertipe data kategorikal, maka penanganan missing valuenya menggunakan nilai modus
# value_counts() => menghitung jumlah nilai (value)
movies['genre'].value_counts()
```

```python
# Karena genre yang paling banyak adalah 'Drama', maka isi missing valuenya dengan 'Drama'
movies.fillna({'genre':'Drama'}, inplace=True)
movies.isna().sum()
```

#### Penanganan 'rating'

Kita akan ubah beberapa rentang rating agar lebih efisien:
*   'SU' (All Ages)
*   '13+' = 'R' = 'PG-13' = 'TV-14'
*   '17+' = 'TV-MA' = 'D' = '21+'

Lalu untuk missing values, akan diisi menjadi 'Unrated'

```python
movies.fillna({'rating':'Unrated'}, inplace=True)
movies.loc[movies['rating'] == 'Unrated']
```

```python
movies['rating'] = movies['rating'].replace({
    "Not Rated": "Unrated",
    "R": "13+",
    "PG-13": "13+",
    "TV-14": "13+",
    "TV-MA": "17+",
    "D": "17+",
    "21+": "17+"
})
movies['rating'].value_counts()
```

#### Penanganan 'directors'

```python
# Penangan pada directors kita bisa langsung hapus saja
movies.dropna(subset=['directors'], inplace=True)
movies.isna().sum()
```

#### Penanganan 'runtime'

```python
# Untuk runtime, kita akan menggunakan modus sebagai pengisi missing values
movies['runtime'].value_counts()
```

```python
movies.fillna({'runtime': 90}, inplace=True)
movies.isna().sum()
```

```python
# Menghapus kata 'min' dan mengubahnya menjadi tipe data integer
movies['runtime'] = movies['runtime'].str.split(pat=" ").str[0]
movies['runtime']
```

### Explore the Data

Movies by Year

```python
f, ax = plt.subplots(figsize=(9, 5))
sns.countplot(x="year", data=movies)

for label in ax.get_xticklabels():
    if int(label.get_text()) % 10 == 0: # filter label tahun per 10 tahun
        label.set_visible(True)
    else:
        label.set_visible(False)
```

Movies by Language

```python
movies["languages"].value_counts().to_frame().rename(columns={"languages" : "total movies"})
```

Movies by TV Ratings

```python
plt.figure(num = None, dpi = 75)
valueCounts = movies["rating"].value_counts()

valueCounts.plot.bar(color='blue')
plt.xlabel('rating')
plt.ylabel('counts')
plt.xticks(rotation=0)
plt.show()
```

Top Movies by User Rating

```python
top_movies = movies.sort_values('users_rating', ascending=False).head(10)
top_movies.iloc[:, :6]
```

Top Directors by average rating of movies

```python
top_directors = movies[["directors", "users_rating"]].groupby(["directors"]).mean()
top_directors.sort_values('users_rating', ascending=False).rename(columns={"users_rating": "average_rating"})[:10]
```
